support: 
S{item_set} = #transactions with item set/ #transaction

confidence:
Conf(A => C) = #txn with both A & c/# txn with A
S({A,C})/S(A)


lift:
How much likely:

L({A => C}) = S({A,C}) / S(C)*S(A)

```python
from efficient_apriori import apriori
import pandas as pd
import numpy as np
from utils.utils import clean_up
```

```python
df = pd.read_csv('data/csv/original/sales2022.csv',dtype={'Scan Code':str})
df = clean_up(df)
df
```

```python
transaction_wise = df.groupby(['Tran ID']).agg({'Description':tuple})
```

```python
transactions = list(transaction_wise['Description'])
transactions
```

```python

item_sets,rules = apriori(transactions,min_support=0.001,min_confidence = 0.001)
print(rules)
```

```python
rules_rhs = filter(lambda rule: len(rule.lhs) == 1 and len(rule.rhs) == 1, rules)
for rule in sorted(rules_rhs, key=lambda rule: rule.lift):
  print(rule)  # Prints the rule and its confidence, support, lift, ...
```
